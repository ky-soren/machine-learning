```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#load the data
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
#check the keys
cancer.keys()
```

```python
#to dataframe
df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
df.head()
```

```python
#too many features to visualize
#let's break it down to two principal components to visualize in 2d space

from sklearn.preprocessing import StandardScaler
```

```python
#instantiate
scaler = StandardScaler()
```

```python
#fit the scaler
scaler.fit(df)
```

```python
#transform the data
scaled_data = scaler.transform(df)
```

```python
#import PCA

from sklearn.decomposition import PCA
```

```python
#we want two components
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
#transform data into our first Principal Component
x_pca = pca.transform(scaled_data)
```

```python
#before
scaled_data.shape
```

```python
#after
#we've reduced it from 30 to 2 components
x_pca.shape
```

```python
#plotting the 2 components
plt.figure(figsize=(10,8))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'],cmap='plasma')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

```python
#we can see there is a clear separation between malignant and benign tumors
#This is done with only 2 columns of data as compared to the original 30
#Good for compression algorithm as we retain a lot of information while reducing the amount of dimensions to the data
```

```python
#taking a closer look at the components
pca.components_

#each row represents a principal component
#each column relates back to the original features
```

```python
#we can visualize this with a heatmap
df_comp = pd.DataFrame(pca.components_,columns=cancer['feature_names'])
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap='plasma')
```

```python
#Now we have a map of the relationship between the correlation of the various features and the principal components.
#Each principal component here is shown as a row.
#The higher the number or the hotter the color, means it's more correlated to a specific feature in the columns
#And this is what the components represent, combinations of all of these features
#And you can see which specific features are important to one of the principal compnent versus the other.

#Done
```
